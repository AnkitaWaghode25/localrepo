```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
df.columns
df.rename(columns={"per capita income (US$)":"price"})
```

```python
df
```

```python
#to replace the space in clm name with  underscore
df.columns=df.columns.str.replace(" ","_")
```

```python
df
#cl=df.per_capita_income_(US$)
```

df

```python
plt.xlabel("year")
plt.ylabel("Per_capita_income_(US$)")
plt.scatter(df['year'] , df['per_capita_income_(US$)'], color='red',marker="*")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per_capita_income_(US$)'])
```

```python
reg.predict(np.array(2020).reshape(-1,1))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python

predicted_value=reg.coef_ * 3300 + reg.intercept_
print(predicted_value)
```
